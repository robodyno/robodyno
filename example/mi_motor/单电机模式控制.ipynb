# 电机模式 Demo

## 初始化电机

```python
from robodyno.components import MiMotor
from robodyno.interfaces import CanBus

can = CanBus()
motor = MiMotor(can, 0x7F)
```

## 使能

```python
motor.enable()
```

## 位置模式

```python
motor.init_pos()
```

```python
from ipywidgets import interact
motor.position_mode()
@interact(pos=(-3.14, 3.14, 0.01))
def demo(pos = 0):
    motor.set_pos(pos)
```

## 速度模式

```python
from ipywidgets import interact
motor.velocity_mode()
@interact(vel=(-3.14, 3.14, 0.01))
def demo(vel = 0):
    motor.set_vel(vel)
```

## 力矩模式

```python
from ipywidgets import interact
motor.torque_mode()
@interact(torque=(-1, 1, 0.01))
def demo(torque = 0):
    motor.set_torque(torque)
```

## 读取电机数据

```python
from ipywidgets import interact
@interact(data = ['位置', '速度', '力矩', '电压', '温度', '模式'])
def demo(data):
    if data == '位置':
        print(motor.get_pos(1))
    elif data == '速度':
        print(motor.get_vel(1))
    elif data == '力矩':
        print(motor.get_torque(1))
    elif data == '电压':
        print(motor.get_voltage(1))
    elif data == '温度':
        print(motor.get_temperature(1))
    elif data == '模式':
        print(motor.get_mode(1))
```

## 失能

```python
motor.disable()
```

## 断开CAN总线

```python
can.disconnect()
```
